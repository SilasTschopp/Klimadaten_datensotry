```python
import matplotlib.pyplot as plt
import numpy as np

# Daten
länder = ['Frankreich', 'Deutschland', 'Schweiz', 'Österreich']
schadholzmenge = [139.6, 30, 12.1, 0.4]
jahresnutzung = [43, 39, 4, 14]
#test

# Erstellen des Plots
fig, ax = plt.subplots(figsize=(12, 8))

# Balkendiagramme für Schadholzmenge und Jahresnutzung
bar_width = 0.35
index = np.arange(len(länder))
schadholzmenge_bars = ax.bar(index, schadholzmenge, bar_width, label='Schadholzmenge (Mio. m3)', color='lightblue')
jahresnutzung_bars = ax.bar(index + bar_width, jahresnutzung, bar_width, label='Jahresnutzung (Mio. m3)', color='orange', alpha=0.7)

# Achsenbeschriftungen und Legende
ax.set_ylabel('Menge (Mio. m3)')
ax.set_xlabel('Länder')
ax.set_title('Schadholzmenge und Jahresnutzung nach Land')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(länder, rotation=45, ha='right')
ax.legend()

# Y-Achsen Schritte festlegen
plt.yticks(np.arange(0, max(max(schadholzmenge), max(jahresnutzung)) + 20, 20))

# Anzeigen des Plots
plt.tight_layout()
plt.show()
```
